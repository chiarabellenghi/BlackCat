```python
from dogma.dogaccess import DogAccess
```

```python
dogs = {}
ids = ["140","141","142","143","150","152","154"]

for id in ids:
    dogs[id] = DogAccess(dog_ip=f"10.1.1.{id}", verbose=True)
```

```python
for id in ids[:4]:
    dogs[id].write(dog_address=0xff000000, dog_registers=[34], dog_data=[0xfe000001])
    dogs[id].write(dog_address=0xff000000, dog_registers=[32], dog_data=[int(id)])
    print()


dogs[ids[4]].write(dog_address=0xff000000, dog_registers=[34], dog_data=[0xfe800002])
dogs[ids[4]].write(dog_address=0xff000000, dog_registers=[32], dog_data=[int(id)])

for id in ids[5:]:
    dogs[id].write(dog_address=0xff000000, dog_registers=[34], dog_data=[0xfe400002])
    dogs[id].write(dog_address=0xff000000, dog_registers=[32], dog_data=[int(id)])
    print()
```
